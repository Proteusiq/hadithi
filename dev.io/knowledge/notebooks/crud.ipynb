```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neighbors import NearestNeighbors
```

```python
text = ["Data Scientist | AI & Machine Learning Engineer | Physicist",
        "Senior Data Scientist @NTT Data / Machine Learning & MLOps / Biophysicist turned Data Scientist",
        "Senior Data Scientist & Tech Lead | Machine Learning & MLOps",
        "Data Scientist & AI Engineer @ NTT Data | MSc Computer Science",
        "Lead AI Solution Architect @ NTT DATA | Teaching AI & ML @ CBS",
        "üêâ Principal Data Scientist | Director of Transformation-Lab", 
       ]

index = ["Mette", "Ulrik", "Shahnoza", "Manuel", "Nicolai", "Prayson",]
```

```python
vectorizer = CountVectorizer(analyzer="word")
docs = vectorizer.fit_transform(text)
```

```python
data = pd.DataFrame.sparse.from_spmatrix(docs, 
                                  columns=vectorizer.get_feature_names_out(),
                                  index=index)

data
```

```python
knn = NearestNeighbors(metric="cosine")
```

```python
knn.fit(data.values)
```

```python
query = "Who is from CBS"
query_mat = pd.DataFrame.sparse.from_spmatrix(vectorizer.transform([query]), 
                                  columns=vectorizer.get_feature_names_out(),
                                  )
query_mat
```

```python
[score], [index] = knn.kneighbors(X=query_mat.values, n_neighbors=1)
data.iloc[index]
```
